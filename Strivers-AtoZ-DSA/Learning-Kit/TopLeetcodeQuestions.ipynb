## **1. Two Sum**

* Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

* You may assume that each input would have exactly one solution, and you may not use the same element twice.

* You can return the answer in any order.

**Example 1:**

Input: nums = [2,7,11,15], target = 9

Output: [0,1]

Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

**Example 2:**

Input: nums = [3,2,4], target = 6

Output: [1,2]

**Example 3:**

Input: nums = [3,3], target = 6

Output: [0,1]
 

**Constraints:**

2 <= nums.length <= 104
-109 <= nums[i] <= 109
-109 <= target <= 109
Only one valid answer exists.
 

**Follow-up:** Can you come up with an algorithm that is less than O(n2) time complexity?

```java
class Solution {
    public int[] twoSum(int[] nums, int target) {
        Map<Integer, Integer> map = new HashMap<>();
        int n = nums.length;
        for(int i = 0; i<n; i++) {
            int num = nums[i];
            int more = target - num;
            if(map.containsKey(more))
                return new int[] {map.get(more), i};
            map.put(num, i);
        }
        return new int[]{};
    }
}
```

**Explanantion**

* We create a `HashMap` to store the elements of the array along with their indices.
* We iterate through the array `nums`.
* For each element `num` at index `i`, we calculate the complement `more` by subtracting `num` from the `target`.
* We check if the complement `more` exists in the HashMap. If it does, it means we have found two numbers that sum up to the target.
* If the complement exists, we return the indices of the current element `num` and the index of its complement from the HashMap.
* If the complement does not exist in the HashMap, we add the current element num along with its index `i` to the HashMap.
* If no such pair is found after iterating through the entire array, we return an empty array to indicate that no solution exists.

* This approach has a time complexity of `O(n)`, where n is the number of elements in the array, because each element is processed only once. The space complexity is also O(n) due to the HashMap, where n is the number of elements in the array.

## **21. Merge Two Sorted Lists**

* You are given the heads of two sorted linked lists list1 and list2.

* Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.

* Return the head of the merged linked list.

**Example 1:**

Input: list1 = [1,2,4], list2 = [1,3,4]

Output: [1,1,2,3,4,4]

**Example 2:**

Input: list1 = [], list2 = []

Output: []

**Example 3:**

Input: list1 = [], list2 = [0]

Output: [0]
 

**Constraints:**

* The number of nodes in both lists is in the range [0, 50].
-100 <= Node.val <= 100
* Both list1 and list2 are sorted in non-decreasing order.

```java
/**
 * Definition for singly-linked list.
 * public class ListNode {
 *     int val;
 *     ListNode next;
 *     ListNode() {}
 *     ListNode(int val) { this.val = val; }
 *     ListNode(int val, ListNode next) { this.val = val; this.next = next; }
 * }
 */

 class Solution {
    public ListNode mergeTwoLists(ListNode list1, ListNode list2) {
        ListNode dummy = new ListNode(-1);
        ListNode current = dummy;

        while (list1 != null && list2 != null) {
            if (list1.val <= list2.val) {
                current.next = list1;
                list1 = list1.next;
            } else {
                current.next = list2;
                list2 = list2.next;
            }
            current = current.next;
        }
        
        current.next = list1 != null ? list1 : list2;
        
        return dummy.next;
    }
}
```

* We start by defining a `dummy` node named dummy, which serves as the head of the merged list. We also initialize a pointer named `current` to keep track of the last node in the merged list.

* We traverse both input linked lists, `list1` and `list2`, simultaneously using a while loop. The loop continues until either `list1` or `list2` becomes null.

* At each iteration, we compare the values of the current nodes from both lists:

    * If the value of the current node in `list1` is less than or equal to the value of the current node in `list2`, we append the current node from `list1` to the merged list by updating the next pointer of the `current` node in current to point to the current node in `list1`. Then, we move the `list1` pointer to its next node.
    * If the value of the current node in `list1` is greater than the value of the current node in `list2`, we append the current node from list2 to the merged list by updating the next pointer of the `current` node in current to point to the current node in `list2`. Then, we move the `list2` pointer to its next node.
* After the loop, we append the remaining nodes from the non-empty list (either `list1` or `list2`) to the merged list by updating the `next` pointer of the `current` node in current to point to the remaining nodes.

* Finally, we return the head of the merged list, excluding the dummy node

## **287. Find the Duplicate Number**
* Given an array of integers nums containing n + 1 integers where each integer is in the range [1, n] inclusive.

* There is only one repeated number in nums, return this repeated number.

* You must solve the problem without modifying the array nums and uses only constant extra space.

**Example 1:**

Input: nums = [1,3,4,2,2]

Output: 2

**Example 2:**

Input: nums = [3,1,3,4,2]

Output: 3
 

**Constraints:**
1 <= n <= 105
nums.length == n + 1
1 <= nums[i] <= n
* All the integers in nums appear only once except for precisely one integer which appears two or more times.

```java
class Solution {
    public int findDuplicate(int[] nums) {
        int slow = nums[0];
        int fast = nums[0];
        
        do {
            slow = nums[slow]; 
            fast = nums[nums[fast]]; 
        } while (slow != fast);

        slow = nums[0];
        while (slow != fast) {
            slow = nums[slow]; 
            fast = nums[fast]; 
        }
        return slow;
    }
}
```

* We initialize two pointers, `slow` and `fast`, both starting at the first element of the array (`nums[0]`).

* We use a do-while loop to move the pointers:
    * The slow pointer moves one step forward by updating its value to nums[slow].
    * The `fast` pointer moves two steps forward by updating its value to nums[nums[fast]].
* We repeat this process until the `slow` pointer equals the `fast` pointer. This indicates the presence of a cycle in the array.

* After finding the meeting point of the two pointers, we reset one of the pointers (`slow` in this case) to the first element of the array (`nums[0]`).

* We then move both pointers at the same pace (one step at a time) until they meet again. The meeting point is the start of the cycle, which corresponds to the duplicate number in the array.

* Finally, we return the duplicate number found at the meeting point.

* This algorithm has a time complexity of `O(n)` and a space complexity of `O(1)`, where n is the length of the array `nums`. It efficiently detects the duplicate number without modifying the original array and using only constant extra space.
